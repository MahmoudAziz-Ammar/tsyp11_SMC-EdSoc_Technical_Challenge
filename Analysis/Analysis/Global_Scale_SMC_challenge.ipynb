# **Step 4:  Modeling**

#**K-Means**


**K-means clustering is an appropriate method for analyzing the relationship between psychological, physiological, social, and academic factors in student stress for 2  reasons**

#**1-Unsupervised Learning**:
K-means is an unsupervised learning algorithm, In the context of student stress, we may not have clear-cut categories or labels for the different levels or types of academic experiences.

#**2-Grouping and Pattern Recognition**:
This can help identify patterns and relationships between the various factors that contribute to student stress. For instance, we might find that students with poor sleep quality tend to have higher levels of anxiety and academic stress.

```python

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt



```

```python
# Standardisez les donn√©es si n√©cessaire
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Utilisez la m√©thode du coude pour trouver le nombre optimal de clusters
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)

# Tracez la m√©thode du coude pour visualiser le nombre optimal de clusters
plt.plot(range(1, 11), wcss)
plt.title('M√©thode du coude')
plt.xlabel('Nombre de clusters')
plt.ylabel('WCSS')  # Within-Cluster Sum of Squares
plt.show()

```

**=> choosing the best parameter to know the best number   using the elbow method**

```python
# Choisissez le nombre optimal de clusters en fonction de la m√©thode du coude
optimal_k = 2# Remplacez par le nombre optimal que vous avez identifi√©
# Ajoutez les √©tiquettes des clusters √† votre DataFrame
data['cluster'] = kmeans.labels_
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=42)
kmeans.fit(data_scaled)

```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Assume data is your DataFrame


# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
kmeans.fit(data)
# Apply PCA to reduce dimensionality to 2 components


pca = PCA(n_components=2)  # Reduce to 2 components
pca_result = pca.fit_transform(data)

# Add the cluster labels to the PCA result
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df['cluster'] = kmeans.labels_

# Plot the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', hue='cluster', palette='viridis', data=pca_df)
plt.title('K-Means Clustering of  Academic Factors')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.show()
```

**=> Cluster 0, Cluster 1, and Cluster 2 represent three distinct groups of students. Each group is characterized by specific traits and behaviors, as indicated by the average values of the features.**

```python
# Calculate cluster means for each attribute
cluster_means = data[data['cluster'].isin([0, 1, 2])].groupby('cluster').mean()

# Transpose the DataFrame for plotting
cluster_means = cluster_means.T

# Plot the cluster means as bar plots
cluster_means.plot(kind='bar', figsize=(12, 6), colormap='viridis')
plt.title('Moyennes des Caract√©ristiques par Cluster')
plt.xlabel('Caract√©ristique')
plt.ylabel('Valeur Moyenne')
plt.legend(title='Cluster')
plt.show()
```

Based on the analysis using K-Means clustering, we have identified three distinct clusters among the individuals:

###**The first cluster** :
consists of individuals with a high level of depression. This group exhibits characteristics associated with increased levels of depression, which may include feelings of sadness, hopelessness, and a negative outlook on life.

###**The second cluster** :
comprises individuals with high self-esteem. This group displays attributes related to a positive self-perception, including confidence, self-worth, and a strong belief in their abilities and value.

###**The third cluster :**
 represents individuals with moderate levels of depression and self-esteem. This group falls between the extremes of the first two clusters, displaying a more balanced combination of psychological factors.
